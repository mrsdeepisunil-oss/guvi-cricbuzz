```python
import sqlite3

def create_db_and_table(db_name):
    if not db_name:
        raise ValueError("Error: Database name not provided")

    conn = sqlite3.connect(db_name)
    cur = conn.cursor()

    cur.execute("""
    CREATE TABLE IF NOT EXISTS employee (
        emp_id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT,
        age INTEGER,
        department TEXT,
        salary INTEGER
    )
    """)

    conn.commit()
    conn.close()
    print("Database and employee table created successfully")
```

```python
def get_column_names(db_name, table_name):
    if not db_name:
        raise ValueError("Error: Database name not provided")

    conn = sqlite3.connect(db_name)
    cur = conn.cursor()

    cur.execute(f"PRAGMA table_info({table_name})")
    columns = cur.fetchall()

    conn.close()

    return [col[1] for col in columns]   # column names only
```

```python
def get_table_values(db_name, table_name):
    if not db_name:
        raise ValueError("Error: Database name not provided")

    conn = sqlite3.connect(db_name)
    cur = conn.cursor()

    cur.execute(f"SELECT * FROM {table_name}")
    rows = cur.fetchall()

    conn.close()
    return rows
```

```python
def add_employee(db_name, name, age, department, salary):
    if not db_name:
        raise ValueError("Error: Database name not provided")

    conn = sqlite3.connect(db_name)
    cur = conn.cursor()

    cur.execute("""
        INSERT INTO employee(name, age, department, salary)
        VALUES (?, ?, ?, ?)
    """, (name, age, department, salary))

    conn.commit()
    conn.close()
    print("Employee added successfully")
```

```python
db = "company.db"

# Create DB and table
create_db_and_table(db)

# Add employees
add_employee(db, "Ravi", 28, "IT", 40000)
add_employee(db, "Anu", 25, "HR", 35000)

# Get columns
cols = get_column_names(db, "employee")
print("Columns:", cols)

# Get values
data = get_table_values(db, "employee")
print("Data:")
for row in data:
    print(row)
```
