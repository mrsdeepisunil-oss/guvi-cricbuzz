# Q2
Show all cricket matches that were played in the last Few days. Include the match description, both team names, venue name with city, and the match date. Sort by most recent matches first.

```python
import requests
import pandas as pd

url = "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/recent"

headers = {
	"x-rapidapi-key": "87cb73cb75msh7a5d872ecde1a59p1feac0jsne6c94486f484",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

match_data = response.json()
```

```python
print(match_data.keys())
```

```python
match_data['filters']
match_data['typeMatches']
```

```python
from datetime import datetime

matches1 = []

for type_match in match_data.get("typeMatches", []):
    for series_matches in type_match.get("seriesMatches", []):
        series_data = series_matches.get("seriesAdWrapper", {})

        for match in series_data.get("matches", []):
            match_info = match.get("matchInfo", {})

            match_desc = match_info.get("matchDesc")
            team1 = match_info.get("team1", {}).get("teamName")
            team2 = match_info.get("team2", {}).get("teamName")

            venue = f"{match_info.get('venueInfo', {}).get('ground')}, {match_info.get('venueInfo', {}).get('city')}"

            start_date = match_info.get("startDate")  # timestamp in ms

            # Convert timestamp → datetime
            if start_date:
                match_date = datetime.fromtimestamp(int(start_date)/1000)
            else:
                match_date = None

            matches1.append({
                "match_desc": match_desc,
                "team1": team1,
                "team2": team2,
                "venue": venue,
                "date": match_date
            })
```

```python
matches_sorted = sorted(matches1, key=lambda x: x["date"], reverse=True)
```

```python
df = pd.DataFrame(matches_sorted)
print(df.head(10))
```

```python
for m in matches_sorted[:10]:   # show latest 10 matches
    print("Match:", m["match_desc"])
    print("Teams:", m["team1"], "vs", m["team2"])
    print("Venue:", m["venue"])
    print("Date:", m["date"].strftime("%Y-%m-%d %H:%M:%S") if m["date"] else "N/A")
    print("-" * 60)
```

```python
import psycopg2

conn = psycopg2.connect(
    host='localhost',
    database='cricbuzz',
    user='postgres',
    password='deepika@88',
    port=5432
)

cur = conn.cursor()

print("✅ PostgreSQL connected successfully")
```

```python
insert_query = """
INSERT INTO recent_matches 
(match_desc, team1, team2, venue, match_date)
VALUES (%s, %s, %s, %s, %s)
"""

for m in matches_sorted:
    cur.execute(insert_query, (
        m.get("match_desc"),
        m.get("team1"),
        m.get("team2"),
        m.get("venue"),
        m.get("date")
    ))

conn.commit()
print("✅ Data inserted into PostgreSQL successfully")
```

Extract batting data & Bowling data

```python
import pandas as pd
import re

# We will store all player data here
player_performance = []

# Process each inning in the scorecard
for inning in data_56717.get("scorecard", []):
    # --- 1. BATSMAN DATA ---
    for b in inning.get("batsman", []):
        name = b.get("name")
        runs = int(b.get("runs", 0))
        balls = int(b.get("balls", 0))
        sr = float(b.get("strkrate", 0))
        out_desc = b.get("outdec", "")
        
        # Determine if the player is "Out" for the average calculation
        # In a single match, if they are out, average = runs. If not out, average is higher.
        is_out = 1 if out_desc.lower() != "not out" else 0
        bat_avg = runs / is_out if is_out > 0 else runs * 1.5 # Not out bonus
        
        # Batting formula: (runs * 0.01) + (avg * 0.5) + (sr * 0.3)
        bat_pts = (runs * 0.01) + (bat_avg * 0.5) + (sr * 0.3)
        
        # --- 2. FIELDING DATA (Parsed from out_desc) ---
        # We look for the player's name in the dismissal descriptions of OTHER players
        # Note: This logic usually requires checking all innings to see who caught whom.
        
        player_performance.append({
            "player": name,
            "format_score": round(bat_pts, 2),
            "runs": runs,
            "balls": balls,
            "strike": sr,
            "type": "Batsman"
        })

    # --- 3. BOWLER DATA ---
    for bw in inning.get("bowler", []):
        bw_name = bw.get("name")
        wkts = int(bw.get("wickets", 0))
        econ = float(bw.get("economy", 0))
        runs_conceded = int(bw.get("runs", 0))
        
        # Bowling average = runs / wickets
        bw_avg = runs_conceded / wkts if wkts > 0 else 50.0
        
        # Bowling formula: (wkts * 2) + (50 - avg) * 0.5 + (6 - econ) * 2
        bowl_pts = (wkts * 2) + ((50 - bw_avg) * 0.5) + ((6 - econ) * 2)
        
        player_performance.append({
            "player": bw_name,
            "format_score": round(bowl_pts, 2),
            "wickets": wkts,
            "economy": econ,
            "type": "Bowler"
        })

# Create the q2024 DataFrame
q19 = pd.DataFrame(player_performance)

# Rank players by total performance score
q19 = q19.sort_values(by="format_score", ascending=False)

print("### TOP PERFORMER RANKINGS ###")
print(q19[["player", "type", "format_score"]].head(10))
```

```python
import pandas as pd
import re

# Assuming 'data' contains your JSON scorecard structure
fielding_stats = {}

def add_stat(player_name, stat_type):
    player_name = player_name.strip()
    if player_name not in fielding_stats:
        fielding_stats[player_name] = {'catches': 0, 'stumpings': 0, 'run_outs': 0}
    fielding_stats[player_name][stat_type] += 1

# Process each inning in the scorecard
for inning in data_56717['scorecard']:
    for batsman in inning.get('batsman', []):
        out_desc = batsman.get('outdec', "").lower()
        
        # 1. Parse Catches: "c Fielder b Bowler" or "c & b Bowler"
        if out_desc.startswith('c '):
            if '& b' in out_desc: # Caught and Bowled
                fielder = out_desc.split('& b')[1].strip()
            else: # Standard Catch
                # Extract text between 'c ' and ' b '
                match = re.search(r'c (.*?) b ', out_desc)
                if match:
                    fielder = match.group(1)
            add_stat(fielder, 'catches')

        # 2. Parse Stumpings: "st Fielder b Bowler"
        elif out_desc.startswith('st '):
            match = re.search(r'st (.*?) b ', out_desc)
            if match:
                fielder = match.group(1)
                add_stat(fielder, 'stumpings')

        # 3. Parse Run Outs: "run out (Fielder)"
        elif 'run out' in out_desc:
            match = re.search(r'\((.*?)\)', out_desc)
            if match:
                fielder = match.group(1)
                # Note: If multiple fielders are listed (e.g., Fielder1/Fielder2), 
                # you can split by '/' and give both credit.
                for f in fielder.split('/'):
                    add_stat(f, 'run_outs')

# Convert to DataFrame for scoring
df_fielding = pd.DataFrame.from_dict(fielding_stats, orient='index').reset_index()
df_fielding.columns = ['Player', 'Catches', 'Stumpings', 'Run_Outs']

# Calculate Weighted Score
# Catch = 10, Stumping = 15, Run Out = 20
df_fielding['Fielding_Score'] = (
    (df_fielding['Catches'] * 10) + 
    (df_fielding['Stumpings'] * 15) + 
    (df_fielding['Run_Outs'] * 20)
)

# Sort by best performers
df_fielding = df_fielding.sort_values(by='Fielding_Score', ascending=False)
print(df_fielding)
```

```python
import pandas as pd
import numpy as np
import re

def calculate_player_rankings(data_56717):
    """
    Calculates weighted performance scores for all players in the scorecard.
    """
    players = {}

    def get_player(name):
        name = name.strip().title()
        if name not in players:
            players[name] = {
                'bat_runs': 0, 'bat_balls': 0, 'bat_outs': 0,
                'bowl_wickets': 0, 'bowl_runs': 0, 'bowl_inns': 0, 'bowl_econ_sum': 0,
                'f_catches': 0, 'f_stumpings': 0, 'f_runouts': 0
            }
        return players[name]

    # Step 1: Extract Data from Scorecard
    for inning in data_56717.get('scorecard', []):
        # Batting & Fielding Extraction
        for b in inning.get('batsman', []):
            p = get_player(b['name'])
            p['bat_runs'] += int(b.get('runs', 0))
            p['bat_balls'] += int(b.get('balls', 0))
            out_desc = b.get('outdec', '').lower()
            if out_desc and 'not out' not in out_desc:
                p['bat_outs'] += 1
            
            # Parsing Fielding (Catches, Stumpings, Run Outs)
            if out_desc.startswith('c '):
                # Handle Caught & Bowled
                if '& b' in out_desc:
                    fielder_name = out_desc.split('& b')[1].strip()
                else:
                    match = re.search(r'^c (.*?) b ', out_desc)
                    fielder_name = match.group(1).replace('(sub)', '').strip() if match else None
                if fielder_name: get_player(fielder_name)['f_catches'] += 1
            
            elif out_desc.startswith('st '):
                match = re.search(r'^st (.*?) b ', out_desc)
                fielder_name = match.group(1).strip() if match else None
                if fielder_name: get_player(fielder_name)['f_stumpings'] += 1
                
            elif 'run out' in out_desc:
                match = re.search(r'\((.*?)\)', out_desc)
                if match:
                    for f in match.group(1).split('/'):
                        get_player(f.strip())['f_runouts'] += 1

        # Bowling Extraction
        for bowl in inning.get('bowler', []):
            p = get_player(bowl['name'])
            p['bowl_wickets'] += int(bowl.get('wickets', 0))
            p['bowl_runs'] += int(bowl.get('runs', 0))
            p['bowl_econ_sum'] += float(bowl.get('economy', 0))
            p['bowl_inns'] += 1

    # Step 2: Apply Weighted Formulas
    results = []
    for name, s in players.items():
        # Batting Points
        bat_avg = s['bat_runs'] / max(1, s['bat_outs'])
        bat_sr = (s['bat_runs'] / max(1, s['bat_balls'])) * 100
        bat_pts = (s['bat_runs'] * 0.01) + (bat_avg * 0.5) + (bat_sr * 0.3) if s['bat_balls'] > 0 else 0
        
        # Bowling Points
        if s['bowl_inns'] > 0:
            # Handle cases where wickets = 0 for average
            b_avg = s['bowl_runs'] / s['bowl_wickets'] if s['bowl_wickets'] > 0 else 50
            avg_econ = s['bowl_econ_sum'] / s['bowl_inns']
            bowl_pts = (s['bowl_wickets'] * 2) + ((50 - b_avg) * 0.5) + ((6 - avg_econ) * 2)
        else:
            bowl_pts = 0
            
        # Fielding Points
        field_pts = (s['f_catches'] * 10) + (s['f_stumpings'] * 15) + (s['f_runouts'] * 20)
        
        total_score = bat_pts + bowl_pts + field_pts
        
        results.append({
            'Player': name,
            'Batting': round(bat_pts, 2),
            'Bowling': round(bowl_pts, 2),
            'Fielding': round(field_pts, 2),
            'Total Score': round(total_score, 2),
            'format' : 'ODI'
        })

    # Return final DataFrame ranked by Total Score
    return pd.DataFrame(results).sort_values(by='Total Score', ascending=False)

# usage:
df_rankings = calculate_player_rankings(data_56717)
print(df_rankings)
```

```python
import pandas as pd
import requests
import time

# List of player IDs to scrape
player_ids = ['41330', '9647', '21823', '37581', '15350', '24827']
all_players_data = []

# Headers for RapidAPI
headers = {
    "x-rapidapi-key": "095038d962msh9068af945c80bc9p1ad37cjsna03b2a42de54",
    "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

print("Fetching data for all players...")

# OUTER LOOP: Iterate through each Player ID
for pid in player_ids:
    url = f"https://cricbuzz-cricket.p.rapidapi.com/stats/v1/player/{pid}"
    
    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code != 200:
            print(f"⚠ Skip {pid}: Status {response.status_code}")
            continue
            
        json_data = response.json()
        
        # 1. Extract Profile Info (Static for each player)
        base_info = {
            'ID': json_data.get('id'),
            'Name': json_data.get('name'),
            'Role': json_data.get('role'),
            'Team': json_data.get('intlTeam')
        }
        
        # 2. INNER LOOP: Iterate through 'recentBatting' rows for this specific player
        # This extracts the match-by-match scores needed for Question 23
        recent_matches = json_data.get('recentBatting', {}).get('rows', [])
        
        for match in recent_matches:
            match_vals = match.get('values', [])
            
            # Create a combined dictionary of Profile + Match Data
            # Note: Index 1 is usually the 'Score' in the JSON you provided
            row_data = base_info.copy()
            row_data.update({
                'Score': match_vals[1],
                'Opponent': match_vals[2],
                'Format': match_vals[3],
                'Date': match_vals[4]
            })
            all_players_data.append(row_data)
            
        print(f"✓ Processed: {base_info['Name']}")
        time.sleep(1)  # Respect API limits

    except Exception as e:
        print(f"✗ Error on {pid}: {e}")

# STEP 3: Create the final DataFrame
df_all_players = pd.DataFrame(all_players_data)

# Display the top rows
print("\n--- FINAL DATAFRAME ---")
print(df_all_players.head(10))
```
