# Q2
Show all cricket matches that were played in the last Few days. Include the match description, both team names, venue name with city, and the match date. Sort by most recent matches first.

```python
import requests
import pandas as pd

url = "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/recent"

headers = {
	"x-rapidapi-key": "87cb73cb75msh7a5d872ecde1a59p1feac0jsne6c94486f484",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

match_data = response.json()
```

```python
print(match_data.keys())
```

```python
match_data['filters']
match_data['typeMatches']
```

```python
from datetime import datetime

matches1 = []

for type_match in match_data.get("typeMatches", []):
    for series_matches in type_match.get("seriesMatches", []):
        series_data = series_matches.get("seriesAdWrapper", {})

        for match in series_data.get("matches", []):
            match_info = match.get("matchInfo", {})

            match_desc = match_info.get("matchDesc")
            team1 = match_info.get("team1", {}).get("teamName")
            team2 = match_info.get("team2", {}).get("teamName")

            venue = f"{match_info.get('venueInfo', {}).get('ground')}, {match_info.get('venueInfo', {}).get('city')}"

            start_date = match_info.get("startDate")  # timestamp in ms

            # Convert timestamp → datetime
            if start_date:
                match_date = datetime.fromtimestamp(int(start_date)/1000)
            else:
                match_date = None

            matches1.append({
                "match_desc": match_desc,
                "team1": team1,
                "team2": team2,
                "venue": venue,
                "date": match_date
            })
```

```python
matches_sorted = sorted(matches1, key=lambda x: x["date"], reverse=True)
```

```python
df = pd.DataFrame(matches_sorted)
print(df.head(10))
```

```python
for m in matches_sorted[:10]:   # show latest 10 matches
    print("Match:", m["match_desc"])
    print("Teams:", m["team1"], "vs", m["team2"])
    print("Venue:", m["venue"])
    print("Date:", m["date"].strftime("%Y-%m-%d %H:%M:%S") if m["date"] else "N/A")
    print("-" * 60)
```

```python
import psycopg2

conn = psycopg2.connect(
    host='localhost',
    database='cricbuzz',
    user='postgres',
    password='deepika@88',
    port=5432
)

cur = conn.cursor()

print("✅ PostgreSQL connected successfully")
```

```python
insert_query = """
INSERT INTO recent_matches 
(match_desc, team1, team2, venue, match_date)
VALUES (%s, %s, %s, %s, %s)
"""

for m in matches_sorted:
    cur.execute(insert_query, (
        m.get("match_desc"),
        m.get("team1"),
        m.get("team2"),
        m.get("venue"),
        m.get("date")
    ))

conn.commit()
print("✅ Data inserted into PostgreSQL successfully")
```

```python
import pandas as pd
import requests
import time

centuries_list = []

print("Extracting centuries from player career endpoints...")

for pid in top_10_odi['playerID']:
    # FIX 1: Removed the "url = " text from inside the string
    career_url = f"https://cricbuzz-cricket.p.rapidapi.com/stats/v1/player/{pid}/batting"
    
    try:
        res = requests.get(career_url, headers=headers)
        if res.status_code == 200:
            career_data = res.json()
            
            # Get headers to find which column is 'ODI'
            header_list = career_data.get('headers', [])
            try:
                # Find the index of 'ODI' in headers (e.g., index 2)
                odi_index = header_list.index('ODI')
            except ValueError:
                centuries_list.append(0)
                continue
            
            # FIX 2: Navigate the table-style JSON
            # We need to find the dictionary where the first value in 'values' is '100s'
            stats_rows = career_data.get('values', [])
            hundreds = 0
            
            for row in stats_rows:
                row_values = row.get('values', [])
                if row_values[0] == '100s':
                    # Grab the value at the same index where 'ODI' was found
                    hundreds = row_values[odi_index]
                    break
            
            centuries_list.append(hundreds)
            print(f"ID {pid}: {hundreds} Centuries found.")
            
        else:
            print(f"Failed to fetch ID {pid}: Status {res.status_code}")
            centuries_list.append(0)
            
    except Exception as e:
        print(f"Error processing ID {pid}: {e}")
        centuries_list.append(0)
    
    time.sleep(1)

# Add to your dataframe
top_10_odi['Centuries'] = centuries_list
print(top_10_odi[['playerID', 'Player', 'Centuries']])
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import time

# --- STEP 1: Extraction Logic ---
all_matches = []

for type_match in data.get("typeMatches", []):
    for series_matches in type_match.get('seriesMatches', []):
        series_data = series_matches.get('seriesAdWrapper', {})
        s_id = series_data.get('seriesId')
        s_name = series_data.get('seriesName')

        for match in series_data.get('matches', []):
            info = match.get('matchInfo', {})
            venue_info = info.get('venueInfo', {})
            
            all_matches.append({
                "series_id": s_id,
                "series_name": s_name,
                "matchId": info.get('matchId'),
                "match_desc": info.get('matchDesc'),
                "match_format": info.get('matchFormat'),
                "start_ts": info.get('startDate'), 
                "status": info.get('status'),
                "team1": info.get('team1', {}).get('teamName'),
                "team2": info.get('team2', {}).get('teamName'),
                "venue": venue_info.get('ground'),
                "city": venue_info.get('city'),
                "country": venue_info.get('country'), # Pulling what API gives us
            })

# --- STEP 2: Create DataFrame ---
df_matches = pd.DataFrame(all_matches)

# --- STEP 3: Handle Dates ---
# Convert start_ts to a readable start_date for Question 8 filtering
df_matches['start_date'] = pd.to_datetime(pd.to_numeric(df_matches['start_ts'], errors='coerce'), unit='ms')

# --- STEP 4: Fill Missing Countries using Geopy ---
# We only want to search for cities where 'country' is currently empty/NaN
missing_country_mask = df_matches['country'].isna() | (df_matches['country'] == "")
cities_to_lookup = df_matches.loc[missing_country_mask, 'city'].unique()

if len(cities_to_lookup) > 0:
    geolocator = Nominatim(user_agent="my_cricket_analyzer_v2")
    geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

    def get_country_safe(city_name):
        if not city_name or city_name == "null": return "Unknown"
        try:
            location = geocode(city_name, language='en')
            if location:
                return location.address.split(',')[-1].strip()
            return "Unknown"
        except:
            return "Error"

    # Map only the missing ones to save API quota
    city_map = {city: get_country_safe(city) for city in cities_to_lookup}
    df_matches.loc[missing_country_mask, 'country'] = df_matches.loc[missing_country_mask, 'city'].map(city_map)

# --- STEP 5: Final Cleanup & View ---
# We keep start_ts as requested, but also have start_date for filtering
print(df_matches[['series_id', 'series_name', 'matchId', 'match_desc', 'match_format', 
                  'start_ts', 'status', 'team1', 'team2', 'venue', 'city', 'country']].head())
```

```python
import pandas as pd

all_batting_records = []

# If sc1 is your list of innings from the API
for inning in sc1:
    inn_id = inning.get('inningsId') or inning.get('inningsid')
    
    # In many Cricbuzz API versions, players are under 'batTeamDetails' -> 'batsmenData'
    # OR directly under 'batsman' as a list. Let's handle both:
    batsmen = inning.get('batsman') or inning.get('batTeamDetails', {}).get('batsmenData', {}).values()
    
    for pos, b in enumerate(batsmen, start=1):
        # Apply the Q19 '10 Balls Faced' Filter
        balls = int(b.get('balls', 0))
        
        if balls >= 10:
            all_batting_records.append({
                "batsman_name": b.get('name'),
                "runs": int(b.get('runs', 0)),
                "balls": balls,
                "strike_rate": b.get('strkrate'),
                "match_id": 105820 
            })

dfq19 = pd.DataFrame(all_batting_records)
print(f"Success! Found {len(dfq19)} qualifying innings.")
```
