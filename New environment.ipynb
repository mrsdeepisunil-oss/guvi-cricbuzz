```python
import plotly.graph_objects as go
from jupyter_dash import JupyterDash
from dash import dcc, html

fig = go.Figure()

app = JupyterDash(__name__)

app.layout = html.Div([
    dcc.Graph(figure=fig)
])

app.run(mode="external", host="0.0.0.0", port=8050)
```
