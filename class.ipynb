```python
from dash import Dash, html, dcc
import plotly.express as px
import pandas as pd


app = Dash()


# assume you have a "long-form" data frame
# see https://plotly.com/python/px-arguments/ for more options
df = pd.DataFrame({
    "Fruit": ["Apples", "Oranges", "Bananas", "Apples", "Oranges", "Bananas"],
    "Amount": [4, 1, 2, 2, 4, 5],
    "City": ["a", "b", "c", "d", "e", "f"]
})


fig = px.bar(df, x="Fruit",y="Amount",barmode='group')
fig1=px.line(df,x="City",y="Amount")


app.layout = html.Div(children=[
    html.H1(children='Hello Dash'),


    html.Div(children='''
        Dash: A web application framework for your data.
    '''),


    dcc.Graph(
        id='example-graph',
        figure=fig
    ),


    dcc.Graph(
        id='example-graph1',
        figure=fig1
    )
])


try:
  if __name__ == '__main__':
    app.run(jupyter_mode="external", port = 8053)
except:
  print("this is an exception")
```

```python
# Run this app with `python app.py` and
# visit http://127.0.0.1:8050/ in your web browser.
from dash import Dash, html, dcc
import plotly.express as px
import pandas as pd


app = Dash()


colors = {
    'background': '#4FE377',
    'text': '#111111'
}


# assume you have a "long-form" data frame
# see https://plotly.com/python/px-arguments/ for more options
df = pd.DataFrame({
    "Fruit": ["Apples", "Oranges", "Bananas", "Apples", "Oranges", "Bananas"],
    "Amount": [4, 1, 2, 2, 4, 5],
    "City": ["SF", "SF", "SF", "Montreal", "Montreal", "Montreal"]
})


fig = px.bar(df, x="Fruit", y="Amount", color="City", barmode="group")
fig1 = px.bar(df, x="Fruit", y="Amount", color="City", barmode="stack")
fig2 = px.bar(df, x="Fruit", y="Amount", color="City", barmode="relative")


fig.update_layout(
    plot_bgcolor=colors['background'],
    paper_bgcolor="#000000",
    font_color=colors['text']
)


app.layout = html.Div(style={'backgroundColor': colors['background']}, children=[
    html.H1(
        children='Hello Dash',
        style={
            'textAlign': 'center',
            'color': colors['text']
        }
    ),


    html.Div(children='Dash: A web application framework for your data.', style={
        'textAlign': 'center',
        'color': colors['text']
    }),


    dcc.Graph(
        id='example-graph-2',
        figure=fig
    ),
     dcc.Graph(
        id='example-graph-21',
        figure=fig1
    ),
     dcc.Graph(
        id='example-graph-22',
        figure=fig2
    ),


    html.P("new line")
])


try:
  if __name__ == '__main__':
    app.run(jupyter_mode="external", port = 8054)
except:
  print("this is and exception")
```

```python
from dash import Dash, html, dcc
import plotly.express as px
import pandas as pd


app = Dash()


app.layout = html.Div([
    html.Div(children=[
        html.Label('Dropdown'),
        dcc.Dropdown(
            options=[
                {'label': 'Chennai', 'value': 'CHN'},
                {'label': 'Bengaluru', 'value': 'BNG'},
                {'label': 'Hydrebad', 'value': 'HYR'},
                {'label':'Mumbai','value':'MUM'}
            ],
            value='CHN'#default value


        ),


        html.Br(),
        html.Label('Multi-Select Dropdown'),
        dcc.Dropdown(
            options=[
                {'label': 'Chennai', 'value': 'CHN'},
                {'label': 'Bengaluru', 'value': 'BNG'},
                {'label': 'Hydrebad', 'value': 'HYR'}
            ],
            #value=['BNG', 'HYR'],
            placeholder="Select a city",
            multi=True
        ),


        html.Br(),#line break in HTML
        html.Label('Radio Items'),
        dcc.RadioItems(
            options=[
                {'label': 'Chennai', 'value': 'CHN'},
                {'label': 'Bengaluru', 'value': 'BNG'},
                {'label': 'Hydrebad', 'value': 'HYR'}
            ],
            value='CHN'
        ),
    ], style={'padding': 10, 'flex': 1}),


    html.Div(children=[
        html.Label('Checkboxes'),
        dcc.Checklist(
            options=[
                {'label': 'Chennai', 'value': 'CHN'},
                {'label': 'Bengaluru', 'value': 'BNG'},
                {'label': 'Hydrebad', 'value': 'HYR'}
            ],
            value=['BNG', 'CHN']
        ),


        html.Br(),
        html.Label('Text Input'),
        dcc.Input(value='CHN', type='text'),


        html.Br(),
        html.Label('Slider'),
        dcc.Slider(
            min=0,
            max=100,
            step=10,
            #marks={0: 'low', 60:'medium',100: 'high'},
            value=7.5,
        ),
    ], style={'padding': 10, 'flex': 1})
], style={'display': 'flex', 'flex-direction': 'row'})


try:
  if __name__ == '__main__':
    app.run(jupyter_mode="external", port = 8055)
except:
  print("this is and exception")


```
